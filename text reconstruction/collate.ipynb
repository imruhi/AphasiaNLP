# Train dataset

```python
import pandas as pd
from nltk.tokenize import sent_tokenize
```

```python
dataf1 = "../text generation/datafiles/generated output/merge_broca.csv"
```

```python
df1 = pd.read_csv(dataf1)#["preprocessed_text"]
```

```python
df1
```

```python
def collate_and_tokenize():
    prompts = []
    for _, row in df1.iterrows():

        # modified = '\n'.join([f"[{index + 1}] {string}" for index, string in enumerate(sent_tokenize(str(row["modified"]).replace('"', r'\"')))])
        modified = str(row["modified"]).replace('"', r'\"')
        # original = '\n'.join([f"[{index + 1}] {string}" for index, string in enumerate(sent_tokenize(str(row["original"]).replace('"', r'\"')))]) 
        original = str(row["original"]).replace('"', r'\"')

        #Merging into one prompt for tokenization and training
        prompt = f"""###System:
Read the sentences provided and answer the corresponding question.
###Sentences:
{modified}
###Question:
Correct the sentences. Give three corrected sentences.
###Answer:
{original}"""
        prompts.append(prompt)
        # print(prompt)
        # break
    return prompts
```

```python
prompts = collate_and_tokenize()
```

```python
pd.DataFrame({"prompt": prompts}).to_csv("data.csv")
```

# Evaluation dataset (generated sentences)

```python
import pandas as pd
from datasets import Dataset
```

```python
dataf1 = "../text generation/datafiles/generated output/merge_broca.csv"
data = pd.read_csv(dataf1)#["preprocessed_text"]
```

```python
data = Dataset.from_pandas(data).train_test_split(test_size=0.3, seed=42)
df1 = data['test']['modified']
```

```python
df1
```

```python
import re
def collate_and_tokenize():
    prompts = []
    for sentence in df1:
        if len(re.findall("[a-zA-Z_]+", sentence)) > 3:
            modified = sentence.replace('"', r'\"')

            prompt = f"""###System:
Read the sentences provided and answer the corresponding question.
###Sentences:
{modified}
###Question:
Correct the sentences. Give three corrected sentences.
###Answer: \n"""
            prompts.append(prompt)
    return prompts
```

```python
prompts = collate_and_tokenize()
```

```python
pd.DataFrame({"prompt": prompts}).to_csv("eval.csv")
```
